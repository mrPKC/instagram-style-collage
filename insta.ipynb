```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
bli = cv2.imread("bottom_left.jpg")
bl = cv2.cvtColor(bli,cv2.COLOR_BGR2RGB)
bl = cv2.resize(bl,(200,200))
bl = cv2.copyMakeBorder(bl,5,10,10,5,cv2.BORDER_CONSTANT)
plt.imshow(bl)
plt.axis("off")
plt.show()
bl.shape
```

```python
bri = cv2.imread("bottom_right.jpg")
br = cv2.cvtColor(bri,cv2.COLOR_BGR2RGB)
br = cv2.resize(br,(200,200))
br = cv2.copyMakeBorder(br,5,10,5,10,cv2.BORDER_CONSTANT)
plt.imshow(br)
plt.axis("off")
plt.show()
br.shape
```

```python
tli = cv2.imread("top_left.jpg",1)
tl = cv2.cvtColor(tli,cv2.COLOR_BGR2RGB)
tl = cv2.resize(tl,(200,200))
tl = cv2.copyMakeBorder(tl,10,5,10,5,cv2.BORDER_CONSTANT)
plt.imshow(tl)
plt.axis("off")
plt.show()
tl.shape
```

```python
tri = cv2.imread("top_right.jpg",1)
tr = cv2.cvtColor(tri,cv2.COLOR_BGR2RGB)
tr = cv2.resize(tr,(200,200))
tr = cv2.copyMakeBorder(tr,10,5,5,10,cv2.BORDER_CONSTANT)
plt.imshow(tr)
plt.axis("off")
plt.show()
tr.shape
```

```python
ci = cv2.imread("center.jpeg")
c = cv2.cvtColor(ci,cv2.COLOR_BGR2RGB)
c = cv2.resize(c,(100,100))
c = cv2.copyMakeBorder(c,10,10,10,10,cv2.BORDER_CONSTANT)
plt.imshow(c)
plt.axis("off")
plt.show()
c.shape
```

```python
final_img_top = cv2.hconcat([tl,tr])
final_img_bottom = cv2.hconcat([bl,br])
final_img = cv2.vconcat([final_img_top,final_img_bottom])
final_img[155:275,155:275] = c
plt.imshow(final_img)
plt.axis("off")
plt.show()
final_img.shape
```

```python
arr = np.array(final_img)
arr=np.reshape(arr, (-1,3),order='C')
df = pd.DataFrame(data=arr,columns=['r','g','b'])
df.to_csv("insta.csv",index=False)
```
